# 1. Важное различие: Claude и FreeX

## Claude
Отдельный AI, которым Андрей пользуется на телефоне.  
Независимый стратегический собеседник.

## FreeX
Объединённая рабочая система, включающая:
- Андрей
- Множество AI (GPT, DeepSeek, Qwen, Kimi, Grok, Gemini, и др.)
- Claude Code на компьютере
- GitHub, документацию, память
- Веб-приложение и рабочий компьютер
- Conference и процессы восстановления

**Claude Code на компьютере = часть FreeX, а не отдельный Claude**

# 2. Почему появился FreeX

Из конкретной личной боли:

> Продолжать мысли, проекты и решения без постоянного начала с нуля.

## Проблемы, которые решал FreeX
- Длинные проекты теряли контекст
- Историю приходилось объяснять заново
- Решения расходились между чатами
- ChatGPT иногда менял принятые решения
- Obsidian требовал много ручного обслуживания
- Notion был чужой системой
- Контекст сложно переносить вручную

# 3. Честное определение FreeX сегодня

**FreeX сегодня:**  
Личная рабочая система Андрея для продолжения мыслей и проектов.  
Объединяет: веб-приложение, заметки, память, AI-агентов, GitHub, документацию и ноутбук.

**Главная формула:**

- **DeepSeek**: память = процесс, не склад

- **Kimi**: продолжающаяся комната; восстановление как функция
- **Grok**: узкий сильный продукт > средний для всех; честная аудитория
- **Gemini**: цифровой экзоскелет; опасность: пользователь обслуживает систему
- **Perplexity**: память = explainable; провенанс важен
- **Copilot**: Project хранит цель, статус, решения, вопросы, шаг
- **NotebookLM**: экзоскелет; агенты имеют компетенции
- **Atlas**: активная поддержка; защита от перегруза
